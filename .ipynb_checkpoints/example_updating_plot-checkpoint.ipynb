```python
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
fig = plt.figure()
```

```python
ax = fig.add_subplot(1, 1, 1) 
```

```python
ax.set_xlabel('X')
ax.set_ylabel('Y');
```

```python
ax.set_xlim(0,1)
ax.set_ylim(0,1);
```

```python
ax.set_title('FIGURE A');
```

```python
for i in range(10):    
    x = np.linspace(0,1,100)
    if ax.lines:
        for line in ax.lines:
            line.set_xdata(x)
            y = np.random.random(size=(100,1))
            line.set_ydata(y)
    else:
        y = np.random.random(size=(100,1))
        ax.plot(x, y, 'b')
    fig.canvas.draw()
    time.sleep(0.1)
```
